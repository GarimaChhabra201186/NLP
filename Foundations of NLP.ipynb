```python
# lexical processing- lexicon - words -> order, grammer,context does not matter
# syntactic Processing- syntax - > order , grammer matters
# semantic processing-semantics ->similarity-> order,grammer,context everything matters
```

```python
# case folding
def case_folding(text):
    return text.lower()
```

```python
# punctuation removal
def remove_punctuation(text):
    return ''.join([char for char in text if char not in string.punctuation])
```

```python
string.punctuation
```

```python
# tokenization
def tokenize(text):
    return nltk.word_tokenize(text)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# stemming 
def stemming(text):
    stemmer = PorterStemmer()
    words = tokenize(text)
    return ' '.join([stemmer.stem(w) for w in words])
```

```python
# lemmatizer 
def lemmatize(text):
    lemmatizer = WordNetLemmatizer()
    words = tokenize(text)
    return ' '.join([lemmatizer.lemmatize(w) for w in words])
```

```python
tweets
```

```python
for t in tweets:
    print("------"*10)
    print("Original : ", t)
    output1 = case_folding(t)
    print(output1)
    output2 = remove_punctuation(output1)
    print(output2)
    output3 = stemming(output2)
    print(output3)
    output4 = lemmatize(output2)
    print(output4) 
```
